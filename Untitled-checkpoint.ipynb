```python
# Documents/Naresh_IT/GenAI/ieee_paper.pdf
import pdfplumber
import camelot
import pandas as pd
import json

pdf_path = "Documents/Naresh_IT/GenAI/ieee_paper.pdf"

data = []  # to store extracted info

with pdfplumber.open(pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        print(f"Processing Page {page_num}...")

        # Extract plain text
        text = page.extract_text()
        if text:
            data.append({
                "page": page_num,
                "type": "text",
                "content": text.strip()
            })

        # Extract tables (optional)
        try:
            tables = camelot.read_pdf(pdf_path, pages=str(page_num))
            for t_id, table in enumerate(tables):
                df = table.df  # dataframe
                data.append({
                    "page": page_num,
                    "type": "table",
                    "content": df.to_dict()
                })
        except Exception as e:
            print(f"No table found on page {page_num} - {e}")
```
